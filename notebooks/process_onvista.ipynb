```python
import pandas as pd
```

```python
def read_onvista_csv(filename):
    df = pd.read_csv(
        filename,
        #index_col=['Datum'],
        parse_dates=['Datum'],
        sep=';',
        decimal=',',
        thousands='.',
        dayfirst=True
    )
    return df

def rename_onvista_cols(df):
    df.columns = [
        "date",
        "open",
        "high",
        "low",
        "close",
        "volume"
    ]
    return df.set_index("date")

def prepare_statworx_df(df_list):
    formatted_dfs = []
    for df in df_list:
        df["change"] = df.close.pct_change()
        df=df.rename(columns={"close":"value"}).drop(["open", "high", "low", "volume"], axis=1)
        formatted_dfs.append(df)
    return pd.concat(formatted_dfs)
```
